Linear regression

```python
X = torch.randn(100, 1)
y = 2 * X + 1 + 0.1 * torch.randn(100, 1)
```

```python
x
```

```python
y
```

```python
w = torch.randn(1, 1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
```

```python
w
```

```python
b
```

```python
learning_rate = 0.1
epochs = 100
```

```python
for epoch in range(epochs):
    # Forward pass (Element-wise multiplication + broadcasting)
    y_pred = X * w + b
    
    # Compute Mean Squared Error (MSE) Loss
    loss = ((y_pred - y) ** 2).mean()
    
    # Backward pass (Compute gradients)
    loss.backward()
    
    # Update weights using gradient descent (no_grad prevents tracking these updates)
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad
        
        # Zero out the gradients for the next epoch
        w.grad.zero_()
        b.grad.zero_()
```

```python

print(f"Trained weight (w): {w.item():.4f}, bias (b): {b.item():.4f}")
print(f"Final Loss: {loss.item():.4f}")
```
