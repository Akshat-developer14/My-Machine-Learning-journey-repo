```python
#input fixed
x = torch.tensor(2.0)

#learnable number
w = torch.tensor(3.0, requires_grad=True)

#correct answer
y_true = torch.tensor(10.0)

#learning rate
lr = 0.1

y_pred = w * x
loss = (y_pred - y_true) ** 2

loss.backward()

print("Before update")
print("w:", w.item())
print("loss", loss.item())
print("gradient", w.grad.item())

# We don't need to track this because it's correcting the value of w not messing the graph
with torch.no_grad(): # do not track
    w -= lr * w.grad

# Reset gradient (clear old gradients)
w.grad.zero_()

print(f"""After update
      w: {w.item()}""")
```

```python
#input fixed
x = torch.tensor(2.0)

#learnable number
w = torch.tensor(0.0, requires_grad=True)

#correct answer
y_true = torch.tensor(10.0)

#learning rate
lr = 0.001
epochs = 10

for epoch in range(epochs):
    # step1 predict
    y_pred = w * x
    # step2 loss
    loss = ((y_pred - y_true) ** 2).mean()

    loss.backward()

    with torch.no_grad():
        w -= lr * w.grad
        w.grad.zero_()

    print(f"Epoch {epoch+1}: w={w.item():.4f}, loss={loss.item():.4f}")
```
