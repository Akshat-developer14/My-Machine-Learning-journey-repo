# Support Vector Classifier (SVC)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Let's create synthetic data points
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(f'''
Confusion Matrix:
{confusion_matrix(y_test, y_pred)}
Classification Report:
{classification_report(y_test, y_pred)}
''')
```

## Doing with non clearly seperable data

```python
X_2, y_2 = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x = pd.DataFrame(X_2)[0], y = pd.DataFrame(X_2)[1], hue=y)
```

```python
rbf = SVC(kernel='rbf')

X_train2, X_test2, y_train2, y_test2 = train_test_split(X_2, y_2, test_size=0.25, random_state=10)

rbf.fit(X_train2, y_train2)
```

```python
svc2 = SVC(kernel='linear')

svc2.fit(X_train2, y_train2)

y_pred_svc2 = svc2.predict(X_test2)

print(f'''
Confusion Matrix:
{confusion_matrix(y_test2, y_pred_svc2)}
Classification Report:
{classification_report(y_test2, y_pred_svc2)}
''')
```

```python
y_pred_rbf = svc.predict(X_test2)

print(f'''
Confusion Matrix:
{confusion_matrix(y_test2, y_pred_rbf)}
Classification Report:
{classification_report(y_test2, y_pred_rbf)}
''')
```
