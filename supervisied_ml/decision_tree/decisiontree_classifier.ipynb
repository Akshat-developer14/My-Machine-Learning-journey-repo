# Decision Tree Classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
# independent features
X = pd.DataFrame(iris['data'], columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
X
```

```python
# Dependent feature
y = iris['target']
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier

treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
# Visualize the Decision tree
from sklearn import tree

plt.figure(figsize=(15, 10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(f'''
Accuracy Score: {accuracy_score(y_test, y_pred)}
Confusion Matrix:
{confusion_matrix(y_test, y_pred)}
Classification Report:
{classification_report(y_test, y_pred)}
''')
```

# Decision Tree Classifier Prepruning And Hyperparameter Tuning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [ 1, 2, 3, 4, 5 ],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV

treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel, param_grid=params, cv=5, scoring='accuracy')
```

```python
import warnings

# Temporarily ignore warnings
with warnings.catch_warnings():
    warnings.filterwarnings("ignore")
    # Also specifically ignore common sklearn warnings if needed
    warnings.simplefilter("ignore", category=UserWarning)
    
    grid.fit(X_train, y_train)
```

```python
grid
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
print(f'''
Accuracy Score: {accuracy_score(y_test, y_pred_grid)}
Confusion Matrix:
{confusion_matrix(y_test, y_pred_grid)}
Classification Report:
{classification_report(y_test, y_pred_grid)}
''')
```
