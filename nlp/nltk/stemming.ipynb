## Stemming 

This is a process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in English. (over, overing, overed) -> over

```python
words = ['running', 'runner', 'runs', 'ran', 'jumping', 'jumps', 'jumped', 'jump', 'eating', 'eats', 'ate', 'eaten', 'fairly', 'fairness', 'sportingly']
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer, RegexpStemmer, SnowballStemmer
```

```python
stemming = PorterStemmer()

for word in words:
    print(word + ' -----> ' + stemming.stem(word))
```

### RegexpStemmer - NLTK has RegexpStemmer which uses regular expression stemmer algorithm to remove the suffixes from the words.
It basically takes a single regular expression and removes any prefix or suffix that matches the regular expression.

```python
reg_stemmer = RegexpStemmer('ing$|ning$|ner$|s$|e$|able$|ed$|ation$|ment$', min=4)
```

```python
for word in words:
    print(word + ' -----> ' + reg_stemmer.stem(word))
```

### Snowball Stemmer

```python
snowball_stemmer = SnowballStemmer('english')

for word in words:
    print(word + ' -----> ' + snowball_stemmer.stem(word))
```
