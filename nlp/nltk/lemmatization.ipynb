## Lemmatization

It is used to convert words into its dictionary form. For example, "running" will be converted to "run", "better" will be converted to "good", etc.

NLTK provides WordNetLemmatizer for lemmatization. It uses WordNet database to find the lemma of a word.

WordNet is a large lexical database of English. Nouns, verbs, adjectives and adverbs are grouped into sets of cognitive synonyms (synsets), each expressing a distinct concept.

It overcomes the problem of stemming by considering the context of the word.

```python
from nltk.stem import WordNetLemmatizer
```

```python
words = ['running', 'runner', 'runs', 'ran', 'jumping', 'jumps', 'jumped', 'jump', 'eating', 'eats', 'ate', 'eaten', 'fairly', 'fairness', 'sportingly', 'history', 'finally', 'finalized']
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()

"""
POS :- (by default = n)
n - noun
v - verb
a - adjective
r - adverb
s - satellite
"""

for word in words:
    print(word + ' -----> ' + lemmatizer.lemmatize(word))
```

- verb (v) is best in my condition as it gives the base form of the word.

```python
for word in words:
    print(word + ' -----> ' + lemmatizer.lemmatize(word, pos='v'))
```

```python
for word in words:
    print(word + ' -----> ' + lemmatizer.lemmatize(word, pos='a'))
```

```python
for word in words:
    print(word + ' -----> ' + lemmatizer.lemmatize(word, pos='r'))
```

```python
for word in words:
    print(word + ' -----> ' + lemmatizer.lemmatize(word, pos='s'))
```
