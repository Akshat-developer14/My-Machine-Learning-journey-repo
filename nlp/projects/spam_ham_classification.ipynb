## Traning a model for spam classification
    - Using BOW, TFIDF, Machine learning algorithm

```python
import pandas as pd
messages = pd.read_csv('../datasets/sms/SMSSpamCollection.txt', sep='\t', names=['label', 'message'])
```

```python
messages.head()
```

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i]) # Just replacing sub words with empty
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
for word in corpus[0:10]:
    print(word)
```

### Create Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))
```

```python
X
```

```python
cv.vocabulary_
```

```python
## Output features
y = pd.get_dummies(messages['label'], drop_first=True)
```

```python
y = messages['label'].map({'spam': 1, 'ham': 0})
```

```python
y.dtypes
```

```python
y
```

```python
y.iloc[:].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
spam_detect_model
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(f"""
Accuracy: {accuracy_score(y_test, y_pred)}
Classification:
{classification_report(y_test, y_pred)}
""")
```
