```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/titanic_dataset_train.csv')
df.head()
```

## 1. Categorical Data

a. Countplot

```python
custom_colors = ['#FF4500', '#1E90FF']
sns.countplot(x = 'Survived', hue='Survived', data=df, palette=custom_colors)
```

```python
df['Survived'].value_counts()
```

b. PieChart

```python
df['Survived'].value_counts().plot.pie(autopct='%.2f')
```

```python
sns.countplot(x = 'Pclass', hue='Survived', data=df, palette=custom_colors)
```

```python
df['Pclass'].value_counts().plot.pie(autopct='%.2f')
```

```python
df['Sex'].value_counts().plot.pie(autopct='%.2f')
```

## 2. Numerical Data

a. Histogram

```python
plt.hist(df['Age'], bins=5)
```

b. Histplot

```python
sns.histplot(df['Age'], kde=True, stat='density')
```

c. Boxplot
- To check outliers in a noisy data.

```python
sns.boxplot(x='Fare', data=df)
```

```python
sns.boxplot(x='Age', data=df)
```

```python
df['Age'].skew()
```
